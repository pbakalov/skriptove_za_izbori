```python
#coding=utf-8
import matplotlib.pylab as plt
# import seaborn as sb
from numpy import arange
# import plotly.graph_objects as go
import pandas as pd

#plt.ion()

from izbori_utils import  get_filtered_results_by_station , load_station_ids


station_ids =  load_station_ids('sections_03.11.2019.txt', 2246) # 2246 = Столична община (22 = код на район, 46 = код на община)

results_by_station, maya, fanduk, x_pos, width, extreme_ids = get_filtered_results_by_station("votes_03.11.2019.txt", station_ids)

print "reading addresses"
adresi = pd.read_excel('./Sekcii-08.09.2019-KM-0.xlsx')
print "addresses read"
#print adresi.head()
sofia_sections = adresi['Област код'.decode('utf-8')] == 22
sofia_sections.head()


```

```python
# izbrani_adresi = pd.DataFrame()
for i, section_id in enumerate(extreme_ids[1:]):
    #region = section_id[:2]
    #municipality = section_id[2:4]
    adm_reg = section_id[4:6]
    section = section_id[6:]
#     print section_id, adresi[sofia_sections  & (adresi['Община код'.decode('utf-8')]==46) \
#                              & (adresi['Район код'.decode('utf-8')]== int(adm_reg)) \
#                              & (adresi['Секция'.decode('utf-8')]  == int(section))]#['Адрес'.decode('utf-8')]
    
#     print section_id
    if i == 0: 
        izbrani_adresi = adresi[sofia_sections  & (adresi['Община код'.decode('utf-8')]==46)\
                                 & (adresi['Район код'.decode('utf-8')]== int(adm_reg)) \
                                 & (adresi['Секция'.decode('utf-8')]  == int(section))]
    else:
        izbrani_adresi = izbrani_adresi.append(adresi[sofia_sections  & (adresi['Община код'.decode('utf-8')]==46)\
                                 & (adresi['Район код'.decode('utf-8')]== int(adm_reg)) \
                                 & (adresi['Секция'.decode('utf-8')]  == int(section))], ignore_index=True)
    


    

izbrani_adresi
```

```python
del izbrani_adresi['Населено място код'.decode('utf-8')]
```

```python

izbrani_adresi['Район код'.decode('utf-8')] = izbrani_adresi['Район код'.decode('utf-8')].apply(lambda x: str(int(x)).zfill(2))
izbrani_adresi['Секция'.decode('utf-8')] = izbrani_adresi['Секция'.decode('utf-8')].apply(lambda x: str(int(x)).zfill(3))
izbrani_adresi =  izbrani_adresi.rename(columns = {'Област текст'.decode('utf-8'):'Област'.decode('utf-8')} )

izbrani_adresi
```
